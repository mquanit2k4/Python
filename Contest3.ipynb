```python
# Bai 15: So Sphenic
from math import*
def check(n):
    res = 0
    for i in range(2, isqrt(n) + 1):
        if n % i == 0:
            count = 0
            while n % i == 0:
                count += 1
                n //= i
                if count > 1:
                    return 0
                else:
                    res += 1
    if n > 1:
        res += 1
    if res == 3:
        return 1
    else:
        return 0
if __name__ == '__main__':
    n = int(input())
    print(check(n))
```

```python
# Bai 16: So Smith
from math import*
def isPrime(n):
    if n < 2:
        return False
    for i in range(2,isqrt(n) + 1):
        if n % i == 0:
            return False
    return True

def sumDigit(n):
    sum = 0 
    while n != 0:
        sum += n % 10
        n //= 10
    return sum

def check(n):
    if isPrime(n) == True:
        return False
    else:
        k = n
        res = 0
        for i in range(2,isqrt(n)+1):
            if n % i == 0:
                while n % i == 0:
                    res += sumDigit(i)
                    n //= i
        if n > 1:
            res += sumDigit(n)
        return res == sumDigit(k)

if __name__ == '__main__':
    n = int(input())
    if check(n):
        print('YES')
    else:
        print('NO')
```

```python
# Bai 23: Tong uoc so
from math import*
def sumDivisor(n):
    k = isqrt(n) + 1
    sum = 0
    for i in range(1,k):
        if n % i == 0:
            sum += i
            if i*i != n:
                sum += n//i
    return sum

if __name__ == '__main__':
    n = int(input())
    print(sumDivisor(n))
```

```python
# Bai 6: So thuan nguyen to
from math import*
def isPrime(n):
    k = isqrt(n) + 1
    if n < 2:
        return False
    for i in range(2,k):
        if n % i == 0:
            return False
    return True

def checkDigit(n):
    k = n
    res = 0
    while n != 0:
        dg = n % 10
        if isPrime(dg):
            res += dg
        else:
            return False
        n //= 10
    return isPrime(res) and isPrime(k)

if __name__ == '__main__':
    a , b = map(int , input().split())
    ans = 0
    for i in range(a, b + 1):
        if checkDigit(i):
            ans += 1
    print(ans)
            
```

```python
# Bai 41: So gan thuan nghich
def checkReverse(n):
    res = 0
    k = n
    while n != 0:
        res = res * 10 + n % 10
        n //= 10
    return res == k

def checkN(n):
    a = n % 10
    k = n
    res = 0
    while n != 0:
        res += 1
        b = n % 10
        n //= 10
    
    ans = (k - (b*pow(10,res - 1) + a)) // 10
    return checkReverse(ans) and (a*2 == b or b*2 == a)

if __name__ == '__main__':
    n = int(input())
    if checkN(n):
        print('YES')
    else:
        print('NO')
```

```python
# Bai 35: UCLN BCNN
def UCLN(a,b):
    while b != 0:
        b , a = a % b , b
    return a

def BCNN(a,b):
    return a*b//UCLN(a,b)
if __name__ == '__main__':
    a , b = map(int , input().split())
    print(UCLN(a,b) , BCNN(a,b))
```

```python
# Bai 32: Thua so nguyen to thu k
from math import*
def thuaSNT(n , k):
    count = 0
    for i in range(2, isqrt(n) + 1):
        if n % i == 0:
            while n % i == 0:
                count += 1
                if count == k:
                    return i
                n //= i
    if n > 1:
        count += 1
    if count == k:
        return n
    else:
        return -1

if __name__ == '__main__':
    n , k = map(int , input().split())
    print(thuaSNT(n,k))
```
